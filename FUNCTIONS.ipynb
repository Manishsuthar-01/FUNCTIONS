**1. What is the difference between a function and a method in Python?**
>**Function:**

>A function is a block of reusable code that performs a specific task. It can be defined anywhere in your code.

>Functions are not tied to any object or class.

>You can call a function without needing an object.




```python
#Example:

def greet(name):
    return f"Hello, {name}!"


greet("Alice")
```

>**Method:**

>A method is a function that is associated with an object or a class.

>Methods are functions that are defined within a class and are meant to be called on instances (objects) of that class.

>They often operate on the data (attributes) of the object and can access/modify the object's state.

>The first parameter of a method is always self (for instance methods), which refers to the instance of the object.



```python
#Example

class Person:
    def __init__(self, name):
        self.name = name

    def greet(self):
        return f"Hello, {self.name}!"


person = Person("Alice")

person.greet()
```

**2. Explain the concept of function arguments and parameters in Python.**
>**Function Parameters:**

>Parameters are variables that are listed in the function definition. They act as placeholders for the values that youâ€™ll pass into the function when calling it.

>When you define a function, the parameters specify what kind of data the function expects to receive.




```python
#Example

def add(a, b):  # 'a' and 'b' are parameters
    return a + b
```

>**Function Arguments:**

>Arguments are the actual values or data you provide to the function when calling it. These correspond to the parameters that are defined in the function.

>Arguments can be passed in several ways: by position, by keyword, or as default values.

>>**Example:**

>>result = add(3, 5)  # 3 and 5 are arguments


>**Types of Arguments:**

>>**Positional Arguments:** Passed in the order the parameters are defined.









```python
def multiply(x, y):
    return x * y

result = multiply(4, 5)
```

```python
def greet(name, greeting="Hello"):
    return f"{greeting}, {name}!"

print(greet("Manish"))
print(greet("Jay", "Hi"))
```

```python
def print_numbers(*args):
    for num in args:
        print(num)

print_numbers(1, 2, 3, 4)
```

>(**kwargs): Accepts a variable number of keyword arguments.

```python
def print_info(**kwargs):
    for key, value in kwargs.items():
        print(f"{key}: {value}")

print_info(name="Manish", age=25)
```

**5. What are iterators in Python and how do they differ from iterables?**
>**Iterable**

>An iterable is any Python object capable of returning its members one at a time. These objects implement the __iter__() method, which returns an iterator. In simpler terms, an iterable is any object you can loop over using a for loop.

>Common examples of iterables are:

>>Lists

>>Tuples

>>Strings

>>Dictionaries (you can iterate over keys, values, or items)

>>Sets

>>Files (each line is an iterable element)


>The key point is that iterables are things you can "iterate over" (like my_list in the example). They provide the structure for iteration, but by themselves, they donâ€™t produce the actual values during iteration.



```python
#Example of an Iterable:

my_list = [1, 2, 3, 4]


for item in my_list:
    print(item)
```

>**Iterator**

>An iterator is an object that represents a stream of data. It produces values from an iterable, one at a time, and tracks the current position. Iterators implement two methods:

>__iter__(): Returns the iterator object itself (i.e., it's both an iterable and an iterator).

>__next__(): Returns the next item from the sequence. When there are no more items left, it raises a StopIteration exception to indicate the end of the iteration.

>The key distinction here is that an iterator keeps track of the current state of the iteration and provides the next value on demand.




```python
#Example of an Iterator:

my_list = [1, 2, 3, 4]


my_iterator = iter(my_list)


print(next(my_iterator))  # Output: 1
print(next(my_iterator))  # Output: 2
print(next(my_iterator))  # Output: 3
print(next(my_iterator))  # Output: 4
```

**6. Explain the concept of generators in Python and how they are defined.**
>Generators in Python are a simple and powerful tool for creating iterators. They allow you to iterate over data without storing the entire data set in memory, making them ideal for working with large data streams or sequences.

>**Key Concepts of Generators:**

>>**Lazy Evaluation:**

>>Generators produce items one at a time and only on demand (when next() is called), which saves memory.

>>**Yield Keyword:**

>>Unlike regular functions that use return, generator functions use the yield keyword to return values.

>>When yield is used, the functionâ€™s state is saved, so execution can resume from that point when called again.

>>**Iterator Protocol:**

>>Generator objects automatically implement the iterator protocol, meaning they have __iter__() and __next__() methods.






```python
#Generator Function (using yield):

def count_up_to(max):
    count = 1
    while count <= max:
        yield count
        count += 1

counter = count_up_to(5)
for num in counter:
    print(num)



```

```python
#Generator Expression (like list comprehensions but with ()):


squares = (x * x for x in range(5))
for square in squares:
    print(square)
```

**7. What are the advantages of using generators over regular functions?**
>Generators in Python offer several key advantages over regular functions, especially when dealing with large datasets, streams of data, or complex iteration logic. Here's a breakdown of the main advantages:

>**Memory Efficiency**

>Generators yield one item at a time instead of storing the entire sequence in memory.

>Great for handling large datasets or infinite sequences.








```python
#Example:

def count_up_to(n):
    for i in range(n):
        yield i  # No list is created

# vs regular function
def count_up_to_list(n):
    return [i for i in range(n)]
```

>**Lazy Evaluation**

>Values are computed only when needed, which can save time and resources.

>This also allows early termination of loops without computing the rest.


>**Improved Performance**

>Since generators avoid the overhead of creating and returning large data structures, they can be faster in certain scenarios.

>**Infinite Sequences**

>You can create generators that produce values forever, which isnâ€™t feasible with regular functions that return lists.


>**implified Code for Iterators**

>Generators abstract away boilerplate iterator code (e.g., maintaining state manually, handling StopIteration, etc.).


```python
#With a generator:


def my_range(n):
    i = 0
    while i < n:
        yield i
        i += 1
```

```python
#Without a generator:

class MyRange:
    def __init__(self, n):
        self.n = n
        self.i = 0
    def __iter__(self):
        return self
    def __next__(self):
        if self.i < self.n:
            val = self.i
            self.i += 1
            return val
        else:
            raise StopIteration
```

>**Pipelining and Composition**

>Generators can be chained together for stream processing, which is common in data processing pipelines.

**4. Write a Python function that checks if a given number is prime or not from 1 to 200.**

```python
def is_prime(n):
    if n < 2 or n > 200:
        return False
    for i in range(2, int(n ** 0.5) + 1):
        if n % i == 0:
            return False
    return True


for num in range(2, 200):     #range between 1 and 200
    if is_prime(num):
        print(f"{num} is a prime number")
    else:
        print(f"{num} is not a prime number")
```

**5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of
terms.**  

```python
class Fibonacci:
    def __init__(self, max_terms):
        self.max_terms = max_terms
        self.count = 0
        self.a = 0
        self.b = 1

    def __iter__(self):
        return self

    def __next__(self):
        if self.count >= self.max_terms:
            raise StopIteration
        if self.count == 0:
            self.count += 1
            return 0
        elif self.count == 1:
            self.count += 1
            return 1
        else:
            fib = self.a + self.b
            self.a = self.b
            self.b = fib
            self.count += 1
            return fib

fib = Fibonacci(10)  # Generate 10 terms
for num in fib:
    print(num)
```

**6. Write a generator function in Python that yields the powers of 2 up to a given exponent.**

```python
def powers_of_two(max_exponent):
    for exp in range(max_exponent + 1):
        yield 2 ** exp

for power in powers_of_two(5):
    print(power)
```

**10. Create a Python program that uses `filter()` to remove all the vowels from a given string**

```python
def remove_vowels(s):
    vowels = 'aeiouAEIOU'
    return ''.join(filter(lambda char: char not in vowels, s))

input_str = "Hello, how are you?"
result = remove_vowels(input_str)
print(result)
```
